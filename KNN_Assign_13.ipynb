**Importing the libraries**

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.linear_model import LogisticRegression
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import cross_val_score
```

**Importing the dataset**

```python
df=pd.read_csv('/content/Social_Network_Ads.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df=df.drop_duplicates()
df.head(4)
```

```python
#encoding gender column
from sklearn import preprocessing
  
le=preprocessing.LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df.head()
```

**Splitting the dataset into the Training set and Test set**

```python
y=df['Purchased']
x=df[['Gender','Age','EstimatedSalary']]
x.head()
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0) 
x_train
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
x_train
```

**Fitting K-NN to the Training set**

```python
knn = KNeighborsClassifier(n_neighbors=4,metric='manhattan')
knn.fit(x_train,y_train)
KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='manhattan',
                     metric_params=None, n_jobs=None, n_neighbors=4, p=2,
                     weights='uniform')
```

**Predicting the Test set results**

```python
from sklearn.metrics import classification_report, accuracy_score
y_pred = knn.predict(x_test)
```

**Making the Confusion Matrix**

```python
print(classification_report(y_test,y_pred))
```

```python
print("Manhattan accuracy:",accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import roc_auc_score
print(" score : ", roc_auc_score(y_test,y_pred)) 
```

```python
#KNN using euclidean distance
from sklearn.neighbors import KNeighborsClassifier
knn1 = KNeighborsClassifier(n_neighbors = 4, metric = 'euclidean')
knn1.fit(x_train, y_train)
```

```python
y_pred1 = knn1.predict(x_test)
print(classification_report(y_test,y_pred1))
```

```python
print("Euclidean accuracy:",accuracy_score(y_test,y_pred1))
```

```python
#AUROC score
print(" AUROC score : ", roc_auc_score(y_test,y_pred1)) 
```
