**Step 1: Importing the libraries**

```python
import pandas as pd
import numpy as np
```

**Step 2: Importing dataset**

```python
df=pd.read_csv("/content/Data.csv")
df.head()
```

**Step 3: Handling the missing data**

```python
df.isnull().sum()
```

```python
df['Salary']=df['Salary'].fillna(df['Salary'].mean())
df.head()
```

```python
df
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df
```

```python
df.isnull().sum()
```

**Step 4: Encoding categorical data**

```python
from sklearn import preprocessing
  
le=preprocessing.LabelEncoder()
df['Purchased']=le.fit_transform(df['Purchased'])
df
```

**Step 5: Creating a dummy variable**

```python
dum_var=pd.get_dummies(df['Country'],prefix='Country')
df=pd.concat([df,dum_var],axis=1)
df
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
x=df[['Age','Salary','Country_France','Country_Germany','Country_Spain']]
y=df[['Purchased']]
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0) # train- test split
```

```python
x_train.head()
```

**Step 7: Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
x_train
```
