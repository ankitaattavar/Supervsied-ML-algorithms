```python

df1 = pd.read_csv('gdrive/My Drive/testKNN.txt',names = ["row","RI", "Na", "Mg", "Al","Si","K","Ca","Ba","Fe","Type"])
df1.head()
```

```python
#Splitting into test and train
y_train = df['Type']
X_train = df.iloc[:,0:10]
X_train
```

```python
y_train.head()
```

```python
#test data
X_test=df1.iloc[:,0:10]
y_test= df1["Type"]
y_test.head()
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
X_train
```

```python
knn = KNeighborsClassifier(n_neighbors=4,metric='manhattan')
knn.fit(X_train,y_train)
KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='manhattan',
                     metric_params=None, n_jobs=None, n_neighbors=4, p=2,
                     weights='uniform')
```

```python
from sklearn.metrics import classification_report, accuracy_score
y_pred = knn.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
print("Manhattan accuracy:",accuracy_score(y_test,y_pred))
```

```python
#Euclidean
from sklearn.neighbors import KNeighborsClassifier
knn1 = KNeighborsClassifier(n_neighbors = 4, metric = 'euclidean')
knn1.fit(X_train, y_train)
```

```python
y_pred1 = knn1.predict(X_test)
print(classification_report(y_test,y_pred1))
```

```python
print("Euclidean accuracy:",accuracy_score(y_test,y_pred1))
```

```python
knn22 = KNeighborsClassifier(n_neighbors=4,metric='manhattan')
knn22.fit(X_train,y_train)
KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='manhattan',
                     metric_params=None, n_jobs=None, n_neighbors=4, p=2,
                     weights='uniform')
```

```python
y_pred22= knn22.predict(X_test)
print(classification_report(y_test,y_pred22))
```

```python
accuracy_score(y_test,y_pred22)
```
